## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Memanggil gambar

```python
image = cv2.imread ("Asset/1.jpg")
```

## Konversi gambar dari RGB ke GRAY

```python
img_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

fig, axs = plt.subplots(1,2, figsize = (10,10))

ax = axs.ravel()

ax[0].imshow(image)
ax[0].set_title("RGB")

ax[1].imshow(img_gray, cmap="gray")
ax[1].set_title("GRAY")
```

## Mendeteksi ambang batas dari tepi

```python
edges = cv2.Canny(image,100,150) 
```

```python
fig, axs = plt.subplots(1,2,figsize = (10,10))
ax = axs.ravel()

ax[0].imshow(img_gray, cmap = "gray")
ax[0].set_title("Gambar Asli")

ax[1].imshow(edges, cmap = "gray")
ax[1].set_title("Gambar yang udah")
```

## Transformasi Hough

```python
lines = cv2.HoughLinesP(edges,1, np.pi/180,225, maxLineGap=65) #function 
image_line = image.copy() #kalau copy yang gray nnti garis ikutan gray
```

## Menggambar garis

```python
for line in lines:
    x1,y1,x2,y2 = line [0]
    cv2.line(image_line, (x1, y1), (x2, y2), (255,0,0), 4)  
```

```python
fig,axs = plt.subplots(1,2,figsize = (10,10))
ax = axs.ravel()

ax[0].imshow(img_gray, cmap = "gray")
ax[0].set_title("Gambar Asli")

ax[1].imshow(image_line, cmap = "gray")
ax[1].set_title("Gambar yang udah")
```
