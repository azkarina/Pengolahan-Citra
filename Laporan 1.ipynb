## RGB to Grayscale with OpenCV

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray') #cmap dipakai agar derajat ke-abu-annya lebih tinggi
#kashrina masyid azka
#202231055
```

## Grayscale to Binary with OpenCV

```python
(thresh, binary) = cv2.threshold(gray, 128, 255,cv2.THRESH_BINARY) #diatas 128 jadi putih
#kashrina masyid azka
#202231055
```

```python
plt.imshow(binary, cmap='gray')
#kashrina masyid azka
#202231055
```
